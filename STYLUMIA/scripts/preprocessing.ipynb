```python
# Function to check for empty containers
def count_empty_containers(series):
    empty_lists = series.apply(lambda x: isinstance(x, list) and len(x) == 0).sum()
    empty_dicts = series.apply(lambda x: isinstance(x, dict) and len(x) == 0).sum()
    empty_tuples = series.apply(lambda x: isinstance(x, tuple) and len(x) == 0).sum()
    return pd.Series({
        '[]': empty_lists,
        '{}': empty_dicts,
        '()': empty_tuples
    })

# 1. Count null values
null_counts = df2.isnull().sum()

# 2. Count empty containers
empty_counts = df2.apply(count_empty_containers).T

# 3. Combine results into a single summary table
summary_table = pd.concat([
    null_counts.rename('Null Values'),
    empty_counts['[]'].rename('Empty Lists []'),
    empty_counts['{}'].rename('Empty Dicts {}'),
    empty_counts['()'].rename('Empty Tuples ()')
], axis=1)

# 4. Add a total row
summary_table.loc['TOTAL'] = summary_table.sum()

# Display results
print("Column-wise Analysis of Empty Values:")
display(summary_table)

# Optional detailed breakdown
print("\nDetailed Breakdown:")
for col in df2.columns:
    col_stats = summary_table.loc[col]
    if any(col_stats > 0):
        print(f"\nColumn: {col}")
        print(f"Null values: {col_stats['Null Values']}")
        print(f"Empty lists: {col_stats['Empty Lists []']}")
        print(f"Empty dicts: {col_stats['Empty Dicts {}']}")
        print(f"Empty tuples: {col_stats['Empty Tuples ()']}")
```

```python
import re

def clean_and_convert(value):
    try:
        # Remove all commas first to fix malformed strings
        clean_str = re.sub(r',\s*', '', str(value))
        
        # Extract the numeric value using regex
        match = re.search(r'\d+\.?\d*', clean_str)
        if match:
            number = float(match.group())
            return int(round(number))  # Round and convert to integer
        return None
    except:
        return None

def format_indian(number):
    """Format number in Indian numbering system (lakhs/crores)"""
    if pd.isna(number):
        return None
    s = str(int(number))
    if len(s) <= 3:
        return s
    result = []
    result.append(s[-3:])  # Last 3 digits
    s = s[:-3]
    while len(s) > 0:
        result.append(s[-2:] if len(s) > 1 else s)  # Next 2 digits
        s = s[:-2]
    return ','.join(reversed(result))

# Clean and extract numeric values
df1['selling_price'] = df1['selling_price'].apply(clean_and_convert)

# Format with Indian numbering system
df1['selling_price'] = df1['selling_price'].apply(format_indian)

# Display results
print(df1['selling_price'].head())
```
